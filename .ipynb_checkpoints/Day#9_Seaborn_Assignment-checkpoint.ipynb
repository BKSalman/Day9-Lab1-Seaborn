## Data Description

- Survival : 0 = No, 1 = Yes
- pclass(Ticket Class) : 1 = 1st, 2 = 2nd, 3 = 3rd
- Sex(Gender) : Male, Female
- Age : Age in years
- SibSp : Number of siblings/spouses abroad the titanic
- Parch : Number of parents/children abrod the titanic
- Ticket : Ticket Number
- Fare : Passenger fare
- Cabin : Cabin Number
- Embarked : Port of Embarkation, C = Cherbourg, Q = Queenstown, S = Southampton

# Import Seaborn

```python
import seaborn as sns
```

# Q9: Draw barplot represents survived people based on sex

```python
sns.barplot(x= df["sex"], y= df["survived"]);
# the average of survived females is higher than the average of survived males
```

# Q10: Draw catplot represents survived people based on embarked

```python
sns.catplot(x="embarked", y="survived", data=df, kind="bar");
# the average of survived people from C is higher than the others
```

# Q11: Draw boxplot represents distribution of male and female based on age and pclass

```python
plt.figure(figsize=(20,10));
sns.boxplot(x= "sex", y= "age",
            hue="pclass",
            data= df
           );
# outliers are more in pclass 3
# pclass 1 has older people compared to the other pclasses
```

# Q12: Draw heatmap represents correlations between sibsp, parch, age, fare, and survived columns

```python
heatmap_df = df[["sibsp", "parch", "age", "fare", "survived"]].corr()
sns.heatmap(data= heatmap_df, annot= True);
```

# Q13: Draw factorplot represents the relation between sibsp and survived columns

```python
sns.catplot(data= df, x="sibsp", y="survived", kind= "point");
```

# Q14: Draw extra insights [Optional]

```python
plt.figure(figsize= (20, 10))
sns.violinplot(data=df, x= "sex", y= "age",
               hue="pclass",
               inner="quart", linewidth=1);
```
